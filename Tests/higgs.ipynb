```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

import xgboost as xgb
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_auc_score

import time
```

```python
higgs = pd.read_csv('../datasets/HIGGS.csv.gz', header=None, nrows=110000)
higgs
```

```python
X_tr, X_ts, y_tr, y_ts = train_test_split(higgs.iloc[:, 1:], higgs.iloc[:, 0], train_size=100000, random_state=0)
```

## Speed + ROC

```python
cls = xgb.XGBClassifier(n_estimators=500,
                        n_jobs=-1,
                        tree_method='exact',
                        max_depth=8,
                        learning_rate=0.1,
                        )

start = time.time()
cls.fit(X_tr, y_tr)
end = time.time()

print((end - start)/500)
print(roc_auc_score(y_ts, cls.predict_proba(X_ts)[:, 1]))
```

```python
cls = xgb.XGBClassifier(n_estimators=500,
                        n_jobs=-1,
                        tree_method='exact',
                        max_depth=8,
                        learning_rate=0.1,
                        colsample_bytree=0.5
                        )

start = time.time()
cls.fit(X_tr, y_tr)
end = time.time()

print((end - start)/500)
print(roc_auc_score(y_ts, cls.predict_proba(X_ts)[:, 1]))
```
